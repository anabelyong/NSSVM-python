```python
import numpy as np
import matplotlib.pyplot as plt

# ours
from solver.funcs import randomData, accuracy, plot2D
from solver.NSSVM import NSSVM
import time 
import psutil
import os 
```

```python
# Get the memory usage before training the model
process = psutil.Process(os.getpid())
memory_before = process.memory_info().rss / (1024 * 1024)  # Memory in MiB

# Data parameters
dim = 0
Ex = ('2D', '3D', 'nD')
m0 = int(50)
n0 = 100

# Generate the data
X, y, tX, ty = randomData(Ex[dim], m0, n0, 0)
[m, n] = X.shape

# SVM parameters
pars = {'C': 0.25, 's0': int(np.ceil(n * (np.log(m/n)) ** 2))}

# Initialize the model and train
model = NSSVM()
out = model.fit(X, y, pars)

# Get the memory usage after training the model
memory_after = process.memory_info().rss / (1024 * 1024)  # Memory in MiB
memory_used = memory_after - memory_before

# Evaluation and logging
acc, _ = accuracy(X, out['w'], y)
tacc, _ = accuracy(tX, out['w'], ty)

print(f"m={m0}")
print(f"Memory Used:                {memory_used:.6f} MiB")
print(f"Training Time:              {out['time']:.3f} sec")
print(f"Training Size:              {m}x{n}")
print(f"Training Accuracy:          {acc*100:.2f}%")
print(f"Testing Size:               {tX.shape[0]}x{n}")
print(f"Testing Accuracy:           {tacc*100:.2f}%")
print(f"Number of Support Vectors:  {out['sv']}")

if hasattr(model, 'alpha'):
    print(f"Support Vector Coefficients, alpha^*: {model.alpha}")

# If 'alpha' is part of the 'out' dictionary
if 'alpha' in out:
    print(f"Support Vector Coefficients, alpha^*: {out['alpha']}")
# Plotting
if Ex[dim] == '2D' and m < 4000:
    fig, axes = plt.subplots(1, 2, figsize=(8, 3))
    fig.subplots_adjust(left=0.12, bottom=0.14, right=0.9, top=0.8, wspace=0.4)

    plot2D(X, y, out['w'], 'NSSVM', acc, ax=axes[0])
    axes[0].set_xlabel('Training data for m=50')

    plot2D(tX, ty, out['w'], 'NSSVM', tacc, ax=axes[1])
    axes[1].set_xlabel('Testing data for m=50')

    plt.show()
```
